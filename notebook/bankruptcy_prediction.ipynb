Machine Learning Bankruptcy Predictions Model 



AIM :- To build a machine learning model that predicts whether a company is likely to go bankrupt based on financial indicators, helping in risk assessment and early warning system

TOOLS: Python, Jupyter Notebook,Scikit-learn,Pandas & NumPy ,Matplotlib/Seaborn (for data visualization)

MODELS USED - Logistic Regression , SVM and Random Forest

TYPE OF MACHINE LEARNING - SUPERVISED LEARNING OF CLASSIFICATION TYPE

```python
import pandas as pd #Data Manipulation and Analysis
import numpy as np #Numerical Python
import os #Access Files and Directory
import matplotlib.pyplot as plt #Plotting the ROC and Confusion Matrix on graph
import seaborn as sns #Statistic Visual Distributiona
from sklearn.metrics import accuracy_score #Computing Accuracy of the Model
from sklearn.model_selection import train_test_split # Spliting the data set on training and testing data
from sklearn.preprocessing import StandardScaler #Pre process the data from kaggle
from sklearn.linear_model import LogisticRegression #Logistic Regression
from sklearn.ensemble import RandomForestClassifier #Random Forest Classifier
from sklearn.svm import SVC #SVM Matrix 
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve #ML evaluation attributes
```

The Cell imports the relevant Packages for - 
1. Reading the data from the directory 
2. Prepossing and Filtering the data and making it available for modelling
3. Model the dataset with Logistic Regression,SVM matrix and Random Forest
4. Split the dataset fir training and testing 
5. Evalauation and ploting the attributes 

```python
df = pd.read_csv(r'C:\Users\DELL\Desktop\ML_Bankruptcy_Prediction\data\data.csv') #Reading the data file from the directory
df.head()
```

```python
df.info()
df.isnull().sum()
df['Bankrupt?'].value_counts()
```

This cell deals with Data Preprocessing (A process to convert teh raw data into meaning format for data modeling)


```python
X = df.drop('Bankrupt?', axis=1)
y = df['Bankrupt?']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

This cell deals with Data Feature scaling and feature selection (A process to scale down/up the numerical feature into a common scale and to select relevant features for modelling purpose)


```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42) #Split Data Set into train train and test data
```

```python
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

This slide deals with Logistic Regression - A supervised machine learning algorithm that accomplishes binary classification tasks by predicting the probability of an outcome, event, or observation.

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

This slide deals with Logistic Regression - A supervised machine learning algorithm that accomplishes binary classification tasks by predicting the probability of an outcome, event, or observation

```python

svm = SVC(probability=True) 
svm.fit(X_train, y_train)
```

This slide deals with SVM (Support Vector Machine) - A supervised machine learning algorithm that classifies data by finding an optimal line or hyperplane that maximizes the distance between each class in an N-dimensional space

```python
models = {
    'Logistic Regression': lr,
    'Random Forest': rf,
    'SVM': svm
}
for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"\n{name} Classification Report:\n", classification_report(y_test, y_pred))
    print(f"{name} Accuracy: {accuracy_score(y_test, y_pred):.2f}")
```

This cell aims to compute the accuracy and the confusion matrix scores for the three models and compare it in a tubulated format.

```python

os.makedirs("assets", exist_ok=True)

model_names = []
accuracies = []

for name, model in models.items():
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    model_names.append(name)
    accuracies.append(acc)

plt.figure(figsize=(8, 6))
plt.bar(model_names, accuracies, color='teal')
plt.ylim(0, 1)
plt.title('Model Accuracy Comparison')
plt.ylabel('Accuracy')
for i, acc in enumerate(accuracies):
    plt.text(i, acc + 0.02, f"{acc:.2f}", ha='center', fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig("assets/accuracy_comparison.png")  # Optional for GitHub README
plt.show()
```

This cell aims to provide a graphical represenation of the Accuracy of the three models used.

```python
from sklearn.metrics import auc

plt.figure(figsize=(8, 6))
for name, model in models.items():
    y_score = model.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend()
plt.tight_layout()
plt.savefig("assets/roc_curve.png")
plt.show()
```

This cell provides a ROC curve of the the 3 models with FPR and TPR on the x and y axis simultenously.
